```python
import numpy as np
import matplotlib.pyplot as plt

# Define b an odd positive integer such that for 0 < a < 1, ab > 1+(3/2)pi
a = 0.5
b = 7

# Number of points to plot and number of n iterations of the fourier series
dx = 100000
n = 200

# Iterations of the fourier series at x = xi
def fourier_series(x, n):
  f_x = 0

  for j in range(1, n):
    f_x += np.cos(b**j*np.pi*x)*(a**j)
  
  return f_x

# Define the list of variables
x = np.linspace(0, 0.6, dx)
weierstrass_func = np.linspace(0, 0.6, dx)

# Iterations of Weierstrass function from x = 0 to x = 0.6
for i in range(dx):
  weierstrass_func[i] = fourier_series(x[i], n)

# Plot the graph
fig, ax = plt.subplots(figsize = (15, 10)) 
ax.plot(x, weierstrass_func,'b-', label = 'Weierstrass Function', markersize = 10)
plt.xlabel(r"$x$")
plt.ylabel(r"$f(x)$")
legend = ax.legend(loc = 'upper left')
plt.grid()
plt.show()
```
